### 第 2 章 马尔可夫决策过程


#### 2.1 马尔可夫性质

随机过程（stochastic process）是概率论的“动力学”部分。概率论的研究对象是静态的随机现象，而随机过程的研究对象是随时间演变的随机现象（例如天气随时间的变化、城市交通随时间的变化）

**定义**：在随机过程中，在给定现在状态和过去所有状态的情况下，其未来状态的条件概率分布仅依赖于当前状态。未来的状态只依赖于当前状态，与历史无关
$$P(S_{t+1} | S_t) = P(S_{t+1} | S_1, S_2, \cdots, S_t)$$
马尔可夫性质可以描述为：给定当前状态，未来的状态与过去的状态是条件独立的。
如果一个随机过程满足马尔可夫性质，那么未来的转移与过去是独立的，只取决于现在。马尔可夫性质是所有马尔可夫过程的基础。

#### 2.2 马尔可夫过程

**定义**：马尔可夫过程是一组具有马尔可夫性质的随机变量序列，其中下一时刻的状态只取决于当前状态。

我们通常用元组 $(\mathcal{S}, \mathcal{P})$ 描述一个马尔可夫过程，其中 $\mathcal{S}$ 是有限数量的状态集合， $\mathcal{P}$ 是状态转移矩阵（state transition matrix）
$$\mathcal{P} = \begin{bmatrix} 
p(s_1 \mid s_1) & p(s_2 \mid s_1) & \cdots & p(s_n \mid s_1) \\
p(s_1 \mid s_2) & p(s_2 \mid s_2) & \cdots & p(s_n \mid s_2) \\
\vdots & \vdots & \ddots & \vdots \\
p(s_1 \mid s_n) & p(s_2 \mid s_n) & \cdots & p(s_n \mid s_n) \\
\end{bmatrix}$$
状态转移矩阵的每一行的和等于 1。

离散时间的马尔可夫过程也叫做马尔可夫链。

#### 2.3 马尔可夫奖励过程（MRP）

**定义**：马尔可夫奖励过程是马尔可夫链加上奖励函数，在马尔可夫奖励过程中，状态转移矩阵和状态都和马尔可夫过程一样，只是多了奖励函数，奖励函数是一个期望，表示我们到达某一个状态的时候，可以获得多大奖励。

MRP 四元组：$(\mathcal{S}, \mathcal{P}, \mathcal{R}, \gamma)$

**奖励函数**：

$$R(s) = \mathbb{E}\left[ R_{t+1} \mid S_t = s\right]$$

**回报（Return）**：

在一个马尔可夫奖励过程中，从第$t$时刻状态开始，直到终止状态时，所有奖励的衰减之和称为回报，公式如下：
$$G_t = r_{t+1} + \gamma r_{t+2} + \gamma^2 r_{t+3} + \cdots +\gamma^{t+k}r_{t+k + 1}+ \cdots + \gamma^{T-t-1} r_{T} = \sum_{k=0}^{T-t-1} \gamma^{k}r_{t+k+1} $$




#### 2.5 马尔可夫决策过程（MDP）

**定义**：在马尔可夫奖励过程（MRP）的基础上加入动作，就得到了马尔可夫决策过程（MDP）。
MDP 与 MRP 非常相像，主要区别为 MDP 中的状态转移函数和奖励函数都比 MRP 多了动作作为自变量。我们不再使用类似 MRP 定义中的状态转移矩阵方式，而是直接表示成了状态转移函数。这样做一是因为此时状态转移与动作也有关，变成了一个三维数组，而不再是一个矩阵（二维数组）；二是因为状态转移函数更具有一般意义，例如，如果状态集合不是有限的，就无法用数组表示，但仍然可以用状态转移函数表示

MDP 五元组：$(\mathcal{S}, \mathcal{A}, \mathcal{P}, \mathcal{R}, \gamma)$
- 状态转移函数：和MRP相比，MDP中的状态转移函数多了和动作相关量。$p(s' \mid s, a)$, 表示在状态$s$下，采取动作$a$后，转移到$s'$的概率。
- 奖励函数：和MRP相比，奖励函数多了和动作相关项。$R(s,a) = \mathbb{E} \left[ R_{t+1} \mid S_t=s, A_t=a\right]$

策略（Policy）：表示当在状态$s$的时候，采取动作$a$的概率，即$\pi(a|s) = P(A_t=a | S_t=s)$

将MDP转换成MRP:

>
>**全概率公式**：
>对于随机变量 $X$ 和条件 $Y$，如果 $Y$ 可以取值 $\{y_1, y_2, \ldots, y_n\}$，则：
>$$P(X) = \sum_{i=1}^{n} P(X \mid Y=y_i) \cdot P(Y=y_i)$$
> 含义：将复杂事件按条件分解，分别计算每种条件下的概率，再加权求和。

状态转移函数：
$$p_{\pi}(s'\mid s)= \sum_{a \in \mathcal{A}}p(a \mid s) \cdot p(s' \mid s, a) = \sum_{a \in \mathcal{A}}\pi(a\mid s)p(s' \mid s,a)$$ 

奖励函数：
$$\begin{aligned} 
R_{\pi}(s) &= \mathbb{E}\left[ R_{t+1} \mid S_t=s\right] \\ 
&= \sum_{r}p(r_{t+1} = r \mid s) \cdot r \\ 
&= \sum_{r} r \cdot \left( \sum_{a \in \mathcal{A} }\left( p(a \mid s) \cdot p(r_{t+1} = r \mid s, a)\right)\right) \\ 
&= \sum_{a \in \mathcal{A}} \pi(a \mid s) \sum_{r} \left( r \cdot p(r_{t+1}=r \mid s, a) \right) \\ 
&= \sum_{a\in \mathcal{A}} \pi(a \mid s) \cdot \mathbb{E} \left[ R_{t+1} \mid s, a\right] \\
&= \sum_{a\in \mathcal{A}} \pi(a \mid s) \cdot R(s,a) 
\end{aligned}$$

> **全期望公式**：
>对于随机变量 $X$ 和条件 $Y$，如果 $Y$ 可以取值 $\{y_1, y_2, \ldots, y_n\}$，则：
>$$\mathbb{E}[X] = \sum_{i=1}^{n} \mathbb{E}[X \mid Y=y_i] \cdot P(Y=y_i)$$
>或者更一般地：
>$$\mathbb{E}[X] = \mathbb{E}_Y[\mathbb{E}_{X|Y}[X \mid Y]]$$
> 含义：分层求平均，加权再汇总。先在每个条件 $Y=y_i$ 下计算 $X$ 的条件期望，再按 $Y$ 的概率分布加权平均。


状态价值函数（在策略 $\pi$ 下）：在策略$\pi$下，当前状态为$s$, 期望获得的收益。$V_\pi(s) = \mathbb{E}_\pi[G_t | S_t = s]$

动作价值函数（Q 函数）：在策略$\pi$下，当前状态为$s$, 采取动作$a$期望的获得的收益：$Q_\pi(s,a) = \mathbb{E}_\pi[G_t | S_t = s, A_t = a]$
- 状态价值函数可以写成动作价值函数的加权和
$$\boxed{\boldsymbol{ V_{\pi}(s) = \sum_{a \in \mathcal{A}} \pi(a \mid s) \cdot Q_{\pi}(s, a)}}$$

> **含义**：当你处于状态 $s$ 时，你的策略 $\pi$ 可能会以不同的概率选择不同的控制指令$a$。因此，当前状态的总价值 $V_\pi(s)$，就是所有可能动作价值 $Q_\pi(s,a)$的期望（加权平均）。

刚进入状态 $s$，还没选动作，接下来要按你的策略$\pi$选动作：每个动作$a$都有$\pi(a \mid s)$的概率被你选中。此时你能预判的总回报，不可能只看某一个动作的价值，必须把所有可能选的动作的价值，按被选中的概率做加权平均，这个平均值，就是你在决策前能拿到的期望回报。

>**证明过程**：
>$$Q_{\pi}(s, a) = \mathbb{E}_\pi[G_t | S_t = s, A_t = a]$$
>目标随机变量 $X \doteq G_t$
>条件 $Y \doteq S_t = s$
>中间变量 $Z \doteq A_t$
>带入全概率公式，得到
>$$\mathbb{E}_\pi[G_t | S_t = s] = \sum_{a \in \mathcal{A}}p(A_t = a \mid S_t=s) \cdot \mathbb{E}_\pi \left[ G_t \mid S_t=s, A_t = a\right]$$

- 动作价值函数和状态价值函数之间存在一定的关系，这个关系可以表示为：
$$\boxed{\boldsymbol{Q_{\pi}(s,a) = R(s, a) + \gamma \sum_{s’ \in \mathcal{S}}p(s’ \mid s, a) \cdot V_\pi(s’)}} $$

> **含义**：此时你处于状态 $s$，并且已经决定执行特定的控制指令 $a$。这个“状态-动作”对的价值，等于必然获得的即时奖励 $R(s,a)$，加上由于系统动力学 $p(s'\mid s,a)$ 导致的未来所有可能到达的新状态 $s'$ 的期望价值。

>**证明过程**：
>将 Q 函数的定义展开，将回报函数的带入
>$$Q_{\pi}(s,a) = \mathbb{E}_{\pi}\left[ R_{t+1} + \gamma \cdot G_{t+1} \mid S_t =s , A_t = a\right] = R(s,a) + >\gamma \cdot \mathbb{E}_{\pi} \left[ G_{t+1} \mid s_t = s, a_t = a\right]$$
>结合 Markov 性质和条件期望引理
>Markov 性质：$\mathbb{E}[G_{t+1} \mid s_{t+1}] = \mathbb{E}[G_{t+1} \mid s_{t+1}, s_t, a_t]$：回报函数 $G_{t+1}$仅仅由$s_{t+1}$决定，和当前状态和动作无关。
>条件期望引理：对于任意的随机变量，$X, Y, Z$: $\mathbb{E}[X \mid Y] = \mathbb{E}[\mathbb{E}[X \mid Z, Y] \mid Y]$
>令：$X= G_{t+1}, Y = (s_t = s, a_t = a), Z=s_{t+1}$
>得到：
>$$ \mathbb{E}_{\pi} \left[ G_{t+1} \mid s_t = s, a_t = a\right] =  \mathbb{E}_{\pi} \left[\mathbb{E} \left[ G_{t+1} >\mid s_{t+1}, s_t = s, a_t = a \right] \mid s_t = s, a_t = a\right] = \mathbb{E}_{\pi} \left[\mathbb{E} \left[ G_{t?>+1} \mid s_{t+1}\right] \mid s_t = s, a_t = a\right]$$

根据状态价值函数的定义
$$\boldsymbol{Q_{\pi}(s,a) = R(s, a) + \gamma \sum_{s’ \in \mathcal{S}}p(s’ \mid s, a) \cdot V_\pi(s’)} $$

#### 2.6 贝尔曼方程（Bellman Equation）

##### 2.6.1 贝尔曼期望方程（对于 MRP）

为了推导贝尔曼方程，需要证明
$$\mathbb{E}\left[ G_{t+1} \mid S_t =s\right] = \mathbb{E} \left[ V(s_{t+1}) \mid s_t\right]$$
令 $G_{t+1} = g'$, $s_{t+1} = s'$
$$\begin{aligned}
\mathbb{E} \left[ V(s_{t+1}) \mid s_t\right] &= \mathbb{E} \left[ \mathbb{E} \left[ G_{t+1} \mid s_{t+1}\right] \mid s_t\right] \\
&= \mathbb{E} \left[ \sum_{g'} g' \cdot  p(g'\mid s')\right] \\
&= \sum_{s'} \sum_{g'} g' \cdot \overbrace{  p(g'\mid s') \cdot p(s' \mid s) }^{\text{根据Markov性质} p(g' \mid s') = p(g' \mid s',s)}\\
&= \sum_{s'} \sum_{g'}\frac{ g' \cdot p(g' \mid s',s)\cdot \overbrace{ p(s' \mid s) \cdot p(s)}^{条件概率：p(s' \mid s) \cdot p(s) = p(s',s)}}{p(s)} \\ 
&= \sum_{s'} \sum_{g'}\frac{ g' \cdot \overbrace{p(g' \mid s',s) \cdot p(s' , s)}^{联合概率链式法则： P(A \mid B,C) P(B,C)=P(A, B,C)}}{p(s)} \\
&= \sum_{s'} \sum_{g'} g'   \cdot \overbrace{\frac{ p(g' , s',s) }{p(s)}}^{条件概率公式：\frac{P(A,B,C)}{P(C) } = P(A, B \mid C)} \\
&=  \sum_{s'} \sum_{g'} g' \cdot p(g' , s'\mid s) \\
&= \sum_{g'} \sum_{s'} g' \cdot p(g' , s'\mid s) \\
&= \sum_{g'}  g' \cdot p(g' \mid s) \\
&= \mathbb{E}\left[ g' \mid s \right] = \mathbb{E}\left[ G_{t+1} \mid s_t \right]

\end{aligned}$$
状态价值函数展开为
$$\boxed{\begin{aligned}
V(s) &= \mathbb{E}\left[\sum_{k=0}^{\infty} \gamma^k r_{t+k} \bigg| S_t = s\right] \\
&= \mathbb{E}\left[ r_{t+1} + \gamma (r_{t+2} + \gamma r_{t+3} + \ldots) \mid S_t =s \right] \\
&=\mathbb{E} \left[ r_{t+1} + \gamma G_{t+1} \mid S_t = s \right] \\
&= R(s) + \gamma \mathbb{E}\left[ V(s_{t+1} )\mid s_t\right] \\
&= R(s) + \gamma \sum_{s' \in S} p(s' \mid s) V(s')
\end{aligned}}$$

MRP的贝尔曼期望方程可以写成矩阵的形式
$$V = R + \gamma \mathcal{P}V$$

展开为
$$\begin{bmatrix} V(s_1) \\ V(s_2) \\\vdots \\ V(s_n)\end{bmatrix} = \begin{bmatrix} R(s_1) \\ R(s_2) \\ \vdots \\ R(s_n)\end{bmatrix} + \gamma \begin{bmatrix} p(s_1 \mid s_1) & p(s_1 \mid s_2) & \cdots & p(s_1 \mid s_n) \\ p(s_2 \mid s_1) & p(s_2 \mid s_2) & \cdots & p(s_2 \mid s_n) \\ \vdots & \vdots & \ddots & \vdots \\ p(s_n \mid s_1) & p(s_n \mid s_2) & \cdots & p(s_n \mid s_n)
\end{bmatrix}\begin{bmatrix} V(s_1) \\ V(s_2) \\\vdots \\ V(s_n)\end{bmatrix}$$

##### 2.6.2 贝尔曼期望方程（对于 MDP）

将MDP过程中状态价值函数和状态-动作价值函数的关系重新组织一起：
$$\boxed{\begin{aligned}Q_{\pi}(s, a) &= R(s, a) + \gamma \sum_{s' \in \mathcal{S}} p(s' \mid s, a) V_{\pi}(s') \\
 V_{\pi}(s) &= \sum_{a \in \mathcal{A}} \pi(a \mid s) \cdot Q_{\pi}(s, a) \end{aligned}} $$

由此可得：

**状态价值函数的贝尔曼方程**：

$$\boxed{\boldsymbol{\begin{aligned} V_{\pi}(s) &= \sum_{a \in \mathcal{A}} \pi(a \mid s) \cdot \underbrace{Q_{\pi}(s, a)}_{\boxed{Q_{\pi}(s, a) = R(s, a) + \gamma \sum_{s' \in \mathcal{S}} p(s' \mid s, a) V_{\pi}(s')}} \\
&= \sum_{a \in \mathcal{A}} \pi(a \mid s) \cdot \left(R(s, a) + \gamma \sum_{s' \in \mathcal{S}} p(s' \mid s, a) \cdot V_{\pi}(s') \right)
\end{aligned}}}$$

> **含义**：当前系统状态的价值，由两层随机性的期望构成：
> 1. 策略的随机性 $\sum_{a \in \mathcal{A}} \pi(a|s)$：你下一步会采取什么控制输入？
> 2. 系统动力学的随机性 $\sum_{s' \in \mathcal{S}} p(s' | s, a)$：在给定的控制输入下，受环境或模型噪声影响，系统实际会转移到哪个具体的下一个状态 $s'$？

**动作价值函数的贝尔曼方程**：

$$\boxed{\begin{aligned} Q_{\pi}(s, a) &= R(s, a) + \gamma \sum_{s' \in \mathcal{S}} p(s' \mid s, a) \underbrace{V_{\pi}(s')}_{\boxed{\sum_{a \in \mathcal{A}} \pi(a \mid s) \cdot Q_{\pi}(s, a) }} \\
&=R(s, a) + \gamma \sum_{s' \in \mathcal{S}}p(s' \mid s, a) \cdot \sum_{a' \in \mathcal{A}} \pi(a' \mid s') \cdot Q_{\pi}(s', a')
\end{aligned}}$$

> **含义**：执行当前动作 $a$ 后，系统状态发生转移进入 $s'$。在 $s'$ 处，控制器将根据策略评估下一个动作 $a'$ 的价值 $Q_\pi(s', a')$。

````python
# ==================================================
# 1. 定义已知环境模型
# ==================================================
S = ["s1", "s2", "s3", "s4", "s5"]  # 状态集合
A = ["保持s1", "前往s1", "前往s2", "前往s3", "前往s4", "前往s5", "概率前往"]  # 动作集合
# 状态转移函数
# mdp 结构： state： action: reward, (next state，transition prob)
MDP = {
    's1': {
        '保持s1' : {'reward' : -1, 'transition': [('s1', 1.0)]},
        '前往s2' : {'reward' :  0, 'transition': [('s2', 1.0)]},
    },
    's2': {
        '前往s1' : {'reward' : -1, 'transition': [('s1', 1.0)]},
        '前往s3' : {'reward' : -2, 'transition': [('s3', 1.0)]},
    },
   's3': {
        '前往s4' : {'reward' : -2, 'transition': [('s4', 1.0)]},
        '前往s5' : {'reward' :  0, 'transition': [('s5', 1.0)]},
    },
   's4': {
        '前往s5' : {'reward' : 10, 'transition': [('s5', 1.0)]},
        '概率前往' : {'reward' :  1, 'transition': [('s2', 0.2),('s3', 0.4),('s4', 0.4)]},
    },

}

# =============================================================
# 2. 核心算法，价值迭代
# $V_{\pi}(s) = \sum_{a \in \mathcal{A}}\pi(a \mid s) \cdot(R(s,a ) +  \gamma \sum_{s \in \mathcal{S}} p(s' \mid s, a) \cdot V(s') )$
# $V_{\pi}(s) = \sum_{a \in \mathcal{A}} Q_\pi(s, a)$
# $Q_\pi(s, a) = R(s, a) + \gamma \sum_{s' \in \mathcal{S}} p(s' \mid s, a) \left(\sum_{a' \in \mathcal{A}} \pi(a' \mid s') \cdot Q_{\pi}(s',a') \right)$
# =============================================================

def evaluate_policy(mdp, policy, gamma=0.5, theta=1e-10):
    """
    给定策略下，所有状态的价值函数
    """
    # 初始化 V
    V = {state: 0 for state in mdp.keys()}
    V['s5'] = 0.0
    iteration = 0
    while True:
        delta = 0.0
        new_V = V.copy()
        for s in mdp.keys():
            v_current = 0.0
            for action, action_info in mdp[s].items():
                action_prob = policy[s].get(action, 0.0)
                if action_prob == 0:
                    continue
                r = action_info['reward']
                expected_future_v = sum(prob * V[next_s] for next_s, prob in action_info['transition'])
                q_value = r + gamma * expected_future_v
                v_current += action_prob * q_value
            new_V[s] = v_current
            delta = max(delta, abs(v_current - V[s]))
        V = new_V
        iteration += 1
        if delta < theta:
            break

    return V, iteration

# =========================================================
# 实验： 对比不同策略
# =========================================================

ramdom_policy = {
    's1': {'保持s1': 0.5, '前往s2': 0.5},
    's2': {'前往s1': 0.5, '前往s3': 0.5},
    's3': {'前往s4': 0.5, '前往s5': 0.5},
    's4': {'前往s5': 0.5, '概率前往': 0.5},
}

# 【策略 B】：贪婪/激进策略（拼命往有奖励的地方冲）
greedy_policy = {
    's1': {'保持s1': 0.1, '前往s2': 0.9}, # 90%概率想去s2
    's2': {'前往s1': 0.1, '前往s3': 0.9}, # 90%概率想去s3
    's3': {'前往s4': 0.9, '前往s5': 0.1}, # 90%概率想去s4（为了s4的10分大奖）
    's4': {'前往s5': 1, '概率前往': 0.} # 90%概率直接拿10分大奖
}

print("=== 正在评估【策略 A：纯随机策略】 ===")
V_random, iters_a = evaluate_policy(mdp=MDP, policy=ramdom_policy)
for state in ['s1', 's2', 's3', 's4']:
    print(f"  V({state}) = {V_random[state]:.4f}")
print(f"(耗时 {iters_a} 轮迭代)\n")

print("=== 正在评估【策略 B：激进贪婪策略】 ===")
V_greedy, iters_b = evaluate_policy(mdp=MDP, policy=greedy_policy)
for state in ['s1', 's2', 's3', 's4']:
    print(f"  V({state}) = {V_greedy[state]:.4f}")
print(f"(耗时 {iters_b} 轮迭代)\n")
````

##### 2.6.4 贝尔曼最优方程

最优状态价值函数：$V^*(s) = \max_\pi V_\pi(s)$

最优动作价值函数：$Q^*(s,a) = \max_\pi Q_\pi(s,a)$

**贝尔曼最优方程**：

$V^*(s)$ 和 $Q^*(s,a)$ 之间的关系：

处于状态 $s$时候的最优价值函数，一定等于在当前状态下，审视了所有可选动作后，挑出的那个能带来最大动作价值的动作所对应的价值：

$$V^*(s) = \max_{a \in \mathcal{A}}Q^{*}(s, a)$$

如果在状态 $s$ 已经被强制执行了动作 $a$，此时的价值函数$Q^*(s,a)$由两部分组成：眼前的奖励$R(s,a)$加上环境状态转移后，未来按照最优策略继续走所能拿到的折扣回报，因为未来是按照最优策略来走的，所以达到下一状态$s'$后的价值就是 $V^*(s')$：

$$Q^*(s,a) = R(s,a) + \gamma \sum_{s' \in \mathcal{S}}p(s' \mid s,a) \cdot V^*(s
')$$
融合上面两个公式，可以得到：

$$\boxed{V^*(s) = \max_{a \in \mathcal{A}}\left( R(s,a) + \gamma \sum_{s' \in \mathcal{S}} p(s' \mid s,a) \cdot V^*(s')\right)}$$

$$\boxed{Q^*(s,a) = R(s,a) + \gamma \sum_{s' \in \mathcal{S}}p(s' \mid s,a) \cdot \max_{a' \in \mathcal{A}}Q^*(s',a')}$$


**最优策略**：

最优策略 $\pi^*$通常是一个确定性策略，只要我们知道了最优动作价值函数$Q^*(s,a)$，最优策略就是直接选取让$Q^*$最大化的动作：

$$\pi^*(a \mid s) = \begin{cases} 1, & \text{if } a = \arg\max_{a \in \mathcal{A}} Q^*(s, a) \\ 0, & \text{otherwise} \end{cases}$$
